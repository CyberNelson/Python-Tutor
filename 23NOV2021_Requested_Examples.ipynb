```python
#This finds the odd numbers in a list/array:
seq=[1,
     1,
     2,2,2,3,3]
def oddy(seq):
    output={}
    for number in seq:
        if number not in output:
            output[number]=1
        else:
            output[number]+=1
    odd_balls=[]
    for number in output:
        if output[number]%2==1:
            odd_balls.append(number)
    return odd_balls
print("odd balls",oddy(seq))
```
