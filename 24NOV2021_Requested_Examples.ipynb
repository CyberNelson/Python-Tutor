```python
def make_tuple():
  """
  Scope:
  -Returns a tuple of the multiples of 10 from 1 to 100 inclusive.
  
  Arguments:
  -None

  Returns:
  -tuple object (A tuple of the multiples of 10 from 1 to 100 inclusive)
  """
  #Create empty tuple object
  output=()
  #Use a for loop to loop through a range from 1 to 101 (the number 101 is excluded) and step 10.
  for number in range(0,101,10):
    #We started the range from 0. 0 is not included in the scope, therefore we need to filter it out.
    if number is not 0:
      #If the number isn't zero then add to the tuple the new number.
      output+=(number,)
  #Return output results of our tuple
  return output
#Call and print results of the tuple created in the method.
print(make_tuple())
```

```python
def mix_cases(s=""):
  """
  Scope:
  -Capitalize every letter with an even index within the string.
  -Lowercase every letter with an odd index within the string.

  Arguments:
  -s (mixed case string)

  Returns:
  -string (mixed case string output)
  """
  #Convert string into a list/array of characters
  #Hint: Behind the scenes, a string object is made up of an array of characters.
  character_array=list(s)
  #Using a for loop, create a varaiable,index, to loop through a range of
  # 0 to the length,"len()", of the array of characters.
  for index in range(len(character_array)):
    #Using modulus (finding remainder of division), if index mod 2 has a
    #remainder of 0, then said index is even, otherwise it's odd.
    if index%2==0:
      #Access the list element using "[index]" and call Class String method upper.
      character_array[index]=character_array[index].upper()
    #Modulus remainder of 1 determines this element as an odd number so case
    #defaults to else.
    else:
      #Access the list element using "[index]" and call Class String method lower.
      character_array[index]=character_array[index].lower()
  #Store a string result from the list/array. .join(<List Object Here>)
  output="".join(character_array)
  #Return results
  return output
#print results of method mix_cases. Passing "ABCDEF ghijkl" to parameter "s"
print(mix_cases("ABCDEF ghijkl"))
```

```python
def user_io():
  """
  Scope:
  -Returns a list of items read from the user until the user enters a empty string.

  Arguments:
  -None

  Returns:
  -list: a list of strings
  """
  #Create an array to store input results.
  output=[]
  #Make an endless while loop statement to continuously loop till "break" -ing out
  #of the loop.
  while True:
    #Prompt user "Enter a string" and store input results to data variable upon enter.
    data=input("Enter a string: ")
    #If the user doesn't enter anything then the stored result is "". That's our
    #que to "break" out of the endless loop according to the scope.
    if "" is data:
      #Discontinue the while loop. Code in loop stops here.
      break
    #If stored string in data has content then append to output list/array.
    output.append(data)
    #Print array/list as prescribed by scope.
    print(output)
  #return output results as presribed in Returns details.
  return output
#Print results of called method user_io
print("Output of method:",user_io())
```

```python
def make_tuple(a=0,b=10):
  """
  Scope:
  -Returns a tuple of the multiples of 10 from a to b inclusive.

  Arguments:
  -a (start number)
  -b (stop number) *Note: add 1 to include number "b" in range. range is exclusive of stop number.

  Returns:
  -tuple: a tuple of the multiples of 10 from a to b inclusive.
  """
  #Add 1 to stop number to include stop number in the range.
  b+=1
  #Create our output tuple that will later be returned results.
  output=()
  #Using a for loop we loop through a range of numbers starting at "a",
  #stopping at "b", and incrementing by 10.
  for number in range(a,b,10):
    #Append number to tuple.
    output+=(number,)
  #Return tuple results.
  return output
#print to terminal the results of the make_tuple method.
print(make_tuple(0,30))
```

```python
def guess_number(n):
  """
  Scope:
  -Reptitively prompt the user for a number between 0 and 100, inclusive.
  -If the user correctly guesses the value of given argument n, print 'WIN' and return.
  -Otherwise, indicate whether the guess was too high or too low.

  Arguments:
  -n (Number to be guessed by user)

  Returns:
  -None
  """
  #Check to see if argument pass is valid for the method to operate within scope.
  if n<0 or n>100:
    #If argument is invalid then alert and return to exit out of the method.
    print("Argument 'n' is not between 0 and 100. Invalid.")
    return None
  #Create an endless while loop to break out of once guess is successful.
  while True:
    #Ask user to Enter a number between 0 and 100. Then cast string into integer class.
    guess=int(input("Enter a number between 0 and 100: "))
    #Create if...elif...else switch statement to determine output in loop iteration.
    #If guess is equal to "n" argument then print "WIN" and break out of while loop.
    if guess==n:
      #Print "WIN" as prescribed in Scope.
      print("WIN")
      #Break out of endless while loop.
      break
    #Test if guess was greater than "n" argument.
    elif guess>n:
      #Print to terminal that the guess was too high.
      print("Guess was too high. Try again.")
    #Default guess through else statement.
    else:
      #Print to terminal that the guess was too low.
      print("Guess was too low. Try again.")
  #Return None since no return argument was prescribed.
  return None
#Call guess_number method with example argument 23.
guess_number(23)
```

```python
def evensandodds(first,last):
  """
  Scope:
  -Print out every even number on a seperate line from parameters first to last, inclusive.
  -Next print out every odd number from first to last, inclusive.

  Arguments:
  -first (Start number, integer)
  -last (Stop number, integer) *note: add 1 to include stop number in range.

  Returns:
  -None
  """
  #Add 1 to "last" number to make that number inclusive.
  last+=1
  #Create a comprehensive array that only adds numbers that return mod 2 of 0
  #from the for loop range.
  even_numbers=[number for number in range(first,last) if number%2==0]
  #Create a comprehensive array that only adds numbers that return mod 2 of 1
  #from the for loop range.
  odd_numbers=[number for number in range(first,last) if number%2==1]
  #Print Even Numbers statement to notify user we are printing out even numbers.
  print("Even Numbers: ")
  #Loop through each even number in list/array.
  for number in even_numbers:
    #print out the number on its own line.
    print(number)
  #Print Odd Numbers statement to notify user we are printing out odd numbers.
  print("Odd Numbers: ")
  #Loop through each odd number in list/array.
  for number in odd_numbers:
    #print out the number on its own line.
    print(number)
  #Return None as prescribed in Returns.
  return None
#Call evensandodds method and pass arguments first=1,stop=10.
evensandodds(first=1,last=10)
```

```python
def strings():
  """
  Scope:
  -Returns a tuple of the following strings
  -Using example (""),"Physics is the universe's operating system"

  Arguments:
  -None

  Returns:
  -tuple: (A tuple of strings)
  """
  #Create a tuple with just an empty string inside.
  output=("",)
  #Append string,"Physics is the universe's operating system",to the tuple.
  output+=("Physics is the universe's operating system",)
  #return output results of tuple.
  return output
#print results of strings method.
print(strings())
```

```python
def reverseit():
  """
  Scope:
  -Read multiple lines from the user on standard input until an empty string is read.
  -Return a list of all these lines without line terminators
  -Each line should be reversed from how it is read in.

  Arguments:
  -None

  Returns:
  -list: (Return a list of all lines with each line reversed)
  """
  #Create an empty array to store the results.
  lines=[]
  #Create an endless while loop till no string is returned from input.
  while True:
    #Prompt user to Enter a string.
    line=input("Enter a string: ")
    #If line variable has nothing inside the string then follow through if statement.
    if line=="":
      #Break out of while loop. Conditions have been met.
      break
    #Break up the entire line string into a list/array of characters.
    line=list(line)
    #Using the reverse method, reverse the entire array/list.
    line.reverse()
    #Append to list/array a string of concatenated array/list.
    lines.append("".join(line))
  #Return array/list of reversed lines as prescribed in Returns.
  return lines
#Print results of called method reverseit.
print(reverseit())
```
