Part B
--Level Goal--
    The password for the next level is stored in a file called readme located in the home directory. Use this password to log into bandit1 using SSH. Whenever you find a password for a level, use SSH (on port 2220) to log into that level and continue the game.

--Commands you may need to solve this level--
    ls, cd, cat, file, du, find
    
--Results--
Password:  boJ9jbbUNNfktd78OOpsqOltutMc3MY1

```python
import paramiko
import pandas as pd
import datetime
class SSH:
    def __init__(self,host,port,username,password):
        self.host=host
        self.port=port
        self.username=username
        self.password=password
        self.connection=self.__connect__()
        self.log={"Time Of Execution":[],"Command":[],"Results Decoded":[],"Results Encoded":[],"Results Size (Bytes)":[]}
    def __connect__(self):
        print("Starting Up SSH Client...")
        ssh=paramiko.SSHClient()
        print("Complete\nAdding Auto Add Policy...")
        ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        print("Complete\nAuthorizing Connection...")
        ssh.connect(self.host,port=self.port,username=self.username,password=self.password)
        print("Complete...\nConnected!\n==========\n\n")
        return ssh
    def Execute(self,command=None):
        print("Executing command: "+str(command))
        ssh_stdin, ssh_stdout, ssh_stderr = self.connection.exec_command(command)
        print("Command Executed...\nLogging Data...")
        Time=str(datetime.datetime.now())
        self.log["Time Of Execution"].append(Time)
        self.log["Command"].append(command)
        try:
            decoded_data=ssh_stdout.read().decode("ascii")
        except:
            decoded_data=None
        try:
            size=len(decoded_data)
        except:
            size=0
        self.log["Results Decoded"].append(decoded_data)
        self.log["Results Encoded"].append(repr(ssh_stdout.read()))
        self.log["Results Size (Bytes)"].append(size)
        print("Data Logged...\nComplete!\n\n--Executed Results--\nTime Of Execution: "+str(Time)+"\nCommand: "+command+"\nResults Size (Bytes): "+str(size)+"\nResults (Decoded): "+str(decoded_data)+"\nResults (Encoded): "+str(repr(ssh_stdout.read()))+"\n==========\n")
        return decoded_data
    def History(self):
        return pd.DataFrame(self.log)

print("Bandit: Level 0 -> Level 1\n==========")
server=SSH("bandit.labs.overthewire.org",2220,"bandit0","bandit0")
print("Password: ",server.Execute("cat readme")+"\n==========\nHistory")
server.History()
```
